# Has_Bled_Skoru_Hesaplama

HAS-BLED skoru, atriyal fibrilasyon (AF) olan hastalarda kanama riskini hesaplamak için kullanılan bir skorlama sistemidir . HAS-BLED skorlaması, AF’li hastalarda kanama için risk tanımlaması yapar. Skorlama sistemi, hastanın hipertansiyon, anormal karaciğer fonksiyonu, anormal böbrek fonksiyonu, inme öyküsü, kanama öyküsü veya anemisi, labil INR, 65 yaş üstü olması, ilaç alımı (ASA/Klopidogrel/NSAİİ) ve alkol alımı gibi faktörleri dikkate alır . Skorlama sistemi, 0-9 arasında bir değer alabilir. Skoru 3 ve üzeri olan hastaların kanama riski açısından anlamlıdır ve dikkatli olunması gerekir .

HAS-BLED skoru hesaplamak için, hastanın hipertansiyon, anormal karaciğer fonksiyonu, anormal böbrek fonksiyonu, inme öyküsü, kanama öyküsü veya anemisi, labil INR, 65 yaş üstü olması, ilaç alımı (ASA/Klopidogrel/NSAİİ) ve alkol alımı gibi faktörlerin varlığına bakılır. Her faktör için 1 puan verilir. Skorlama sistemi, toplam puanı 9’a kadar çıkarabilir. Skoru 3 ve üzeri olan hastaların kanama riski açısından anlamlıdır ve dikkatli olunması gerekir

## has-bled skorunu foksiyon kullanarak yazmak

```python
def has_bled_skoru_hesapla():
    try:
        hastanin_yasi=int(input("Hastanın yaşını giriniz:"))
        hasta_inr_degeri=int(input("Hastanın kanının INR değerini giriniz:"))#% olarak 60 dan küçükse.kandaki inr değeri yaklaşık 1 olmalıdır
        hasta_karaciger_hastalık=input("Hasta karaciğer hastalığı geçirdiyse 'evet' yazın, aksi halde 'hayır' yazın:").lower()
        hasta_inme=input("Hasta daha önce inme geçirdiyse 'evet'yazın, aksi halde ise 'hayır'yazın:").lower()
        hasta_kreatinin = float(input("Hastanın serum kreatinin değerini girin: "))#böbrek hastalığı için belirleyici değer 2.26mg/dL
        hasta_kanama=input("Hasta daha önce kanama geçirdiyse veya kanamaya yakın ise 'evet'yazın,aksi bir durum var ise 'hayır'yazın:").lower()
        hasta_ilac_kullanımı=input("Hastanın kanamaya zemin hazırlayan ilaç kullanımı var ise 'evet'yazın, yok ise 'hayır'yazın:").lower()#aspirin,klopidogrel vb.
        hasta_hipertansiyon = input("Hasta hipertansiyon hastalığına sahipse 'evet' yazın, sahip değil ise 'hayır' yazın: ").lower()
        hasta_alkol_kullanimi = input("Hasta haftada en az bir kez alkol kullanıyorsa 'evet' yazın, kullanmıyorsa 'hayır' yazın: ").lower()

        kritik_deger=3
        skor=0
        if hastanin_yasi>=65:
            skor+=1
        if hasta_inr_degeri<60:
            skor+=1
        if hasta_karaciger_hastalık=='evet':
            skor+=1
        if hasta_inme=='evet':
            skor+=1
        if hasta_kreatinin > 2.26:
            skor+=1
        if hasta_kanama=='evet':
            skor+=1
        if hasta_ilac_kullanımı=='evet':
            skor+=1
        if hasta_hipertansiyon=='evet':
            skor+=1
        if hasta_alkol_kullanimi=='evet':
            skor+=1
            
            
  
        kritik_deger = 3

       
        if skor > kritik_deger:
            print("Kanama riskiniz yüksek! Acil bir uzmanı ile görüşün.")

        
         
        print(f"\nHas Bled Skoru: {skor}")

        
    except ValueError:
        return "Hatalı giriş! Lütfen doğru değerler girin."


has_bled_skoru_hesapla()
        
    
```

## tkinter kullanarak has bled hesaplama

```python
from tkinter import*
from tkinter import messagebox

class HasBledUygulama(Tk):
    def __init__(self):
        super().__init__()
        self.title("Has Bled Skoru ")
        self.geometry("400x300")
        self.menuOlustur()
        self.araclar()
        
   
    
    
    def araclar(self):
        self.yasL=Label(text="Hastanın yaşı:").place(x=20,y=20,width=100)
        self.yasE=Entry().place(x=120,y=20,width=100)
                               
        
        self.kreatininL=Label(text="kreatinin değeri :").place(x=20,y=50,width=100)
        self.kreatininE=Entry().place(x=120,y=50,width=100)
         
        self.inrDeger=Label(text="INR Değeri giriniz:").place(x=20,y=80,width=100)
        self.inrDegerE=Entry().place(x=120,y=80,width=100)
        
        self.var_kanama=BooleanVar()
        self.kanama=Checkbutton(text="Kanama Geçmişi:",variable=self.var_kanama).place(x=20,y=110)
        
        self.var_karaciger_hastaligi=BooleanVar()
        self.karaciğer_hastaligi=Checkbutton(text="Karaciğer Hastalığı",variable=self.var_karaciger_hastaligi).place(x=20,y=140)
        
        self.var_inme=BooleanVar()
        self.inme=Checkbutton(text="İnme Geçirdi",variable=self.var_inme).place(x=20,y=170)
        
       
        self.var_ilac=BooleanVar()
        self.ilac_kullanimi=Checkbutton(text="İlaç kullanımı:",variable=self.var_ilac).place(x=20,y=170)
        
        self.var_hipertansiyon=BooleanVar()
        self.hipertansiyon=Checkbutton(text="Hipertansiyon",variable=self.var_hipertansiyon).place(x=20,y=200)
        
        self.var_alkol=BooleanVar()
        self.alkol=Checkbutton(text="Alkol kullanımı",variable=self.var_alkol).place(x=20,y=230)
        
        self.hesapla_buton =Button(self, text="Hesapla", command=self.has_bled_skoru_hesapla)
        self.hesapla_buton.place(x=20,y=260,width=200)

        
    def aciklama(self):
        self.pencere2 = Toplevel()
        self.pencere2.geometry("500x500")
        self.pencere2.title("Has Bled Skoru hesaplama")
        
        metin="HAS-BLED skoru, atriyal fibrilasyon (AF) olan hastalarda kanama riskini hesaplamak için kullanılan bir skorlama sistemidir.Her faktör için 1 puan verilir. Skorlama sistemi, toplam puanı 9’a kadar çıkarabilir. Skoru 3 ve üzeri olan hastaların kanama riski açısından anlamlıdır ve dikkatli olunması gerekir"
        
        label_aciklama=Label(self.pencere2,text=metin)
        label_aciklama.place(x=0,y=0)
    def ogrenci(self):
        self.pencere3 = Toplevel()
        self.pencere3.title("Öğrenci Bilgileri")

        self.labelogrenci = Label(self.pencere3,text="Ceren Köseoğlu 2110235007").place(x=0,y=0)
        
    def menuOlustur(self):
        self.menuBar=Menu(self)
        self.config(menu=self.menuBar)
        
        self.file=Menu(self.menuBar, tearoff=0)
        self.file.add_command(label="Açıklama",command=self.aciklama)
        self.file.add_command(label="Öğrenci", command=self.ogrenci)
        self.menuBar.add_cascade(label="Giriş",menu=self.file)
        
          
    def has_bled_skoru_hesapla(self):
        try:
            hastanin_yasi=int(self.yasE.get())
            hasta_inr_degeri=int(self.inrDegerE.get())
            hasta_karaciger_hastalık=self.var_karaciger_hastaligi.get()
            hasta_inme=self.var_inme.get()
            hasta_kreatinin = float(self.kreatininE.get())
            hasta_kanama=self.var_kanama.get()
            hasta_ilac_kullanimi=self.var_ilac.get()
            hasta_hipertansiyon =self.var_hipertansiyon.get()
            hasta_alkol_kullanimi = self.var_alkol.get()

        
            self.skor=0
            if hastanin_yasi>=65:
                self.skor+=1
            if hasta_inr_degeri<60:
                self.skor+=1
            if hasta_karaciger_hastalık=='evet':
                self.skor+=1
            if hasta_inme=='evet':
                self.skor+=1
            if hasta_kreatinin > 2.26:
                self.skor+=1
            if hasta_kanama=='evet':
                self.skor+=1
            if hasta_ilac_kullanimi=='evet':
                self.skor+=1
            if hasta_hipertansiyon=='evet':
                self.skor+=1
            if hasta_alkol_kullanimi=='evet':
                self.skor+=1
    
   
        except ValueError:
            messagebox.showerror("Hata", "Hatalı giriş! Lütfen doğru değerler girin.")
            
    def bilgiGoster(self):
        messagebox.showinfo("Has Bled Skoru", f"Has Bled Skoru: {skor}")
  
pencere=HasBledUygulama()
pencere.mainloop()

    
        
```
